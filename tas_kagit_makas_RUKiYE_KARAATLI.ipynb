```python
import random
import time

def tas_kagit_makas_RUKiYE_KARAATLI():
    # Oyun tanıtımının gerçekleştirilmesi
    print("🪨 Taş, 📄 Kağıt, ✂️ Makas Oyununa Hoş Geldiniz!")
    print("Kurallar çok basit: Taş makası kırar ve kazanır, Makas kağıdı keser ve kazanır, Kağıt taşı sarar ve kazanır.")
    print("İlk iki turu kazanan oyunun galibi olur!")
    print("Oyundan çıkmak için 'çık' yazmanız yeterli.\n")

    # Seçeneklerin belirtilmesi 
    my_choices= ["taş", "kağıt", "makas"]
    game_count = 0
    player_wins = 0
    computer_wins = 0

    while True:
        # Her yeni oyun için sayaçlar sıfırlanır
        game_count += 1
        round_count = 0
        player_score = 0
        computer_score = 0

        print(f"\nOyun {game_count}: Başlıyor!")
        
        while player_score < 2 and computer_score < 2:
            round_count += 1
            print(f"\nTur {round_count}:")
            
            # Oyuncu seçimi
            player_choices = input("Lütfen Seçiminizi yapın: 🪨 Taş, 📄 Kağıt, ✂️ Makas? :  ").lower()
            if player_choices == 'çık':
                print("Oyundan çıkılıyor... Hoşçakalın!")
                return
            elif player_choices not in my_choices:
                print("⚠️ Geçersiz seçim! Lütfen taş, kağıt veya makas seçmeyi deneyin.")
                continue
            
            # Bilgisayar seçimi
            computer_choices = random.choice(my_choices)
            print(f"Bilgisayar {computer_choices} seçti.")
            
            # Sonuç belirleme
            if player_choices == computer_choices:
                print("😐 Berabere!")
            elif (player_choices == "taş" and computer_choices == "makas") or \
                 (player_choices == "makas" and computer_choices == "kağıt") or \
                 (player_choices == "kağıt" and computer_choices == "taş"):
                print("🎉 Bu turu siz kazandınız!")
                player_score += 1
            else:
                print("💻 Bilgisayar bu turu kazandı!")
                computer_score += 1
            
            print(f"Durum: Oyuncu {player_score} - {computer_score} Bilgisayar")
        
        # Oyunun kazananının belirlenmesi
        if player_score == 2:
            print("Tebrikler! Oyunu kazandınız! 🎉")
            player_wins += 1
        else:
            print("Maalesef bilgisayar oyunu kazandı. 😔")
            computer_wins += 1
        
        # Devam etme isteği
        play_again = input("Başka bir oyun oynamak ister misiniz? (evet/hayır): ").lower()
        if play_again != 'evet':
            print("Oyunu bitiriyorsunuz. Oynadığınız için teşekkürler!")
            break
        else:
            # Bilgisayarın devam isteği
            computer_play_again = random.choice(["evet", "hayır"])
            print(f"Bilgisayar {computer_play_again} dedi.")
            if computer_play_again != "evet":
                print("Bilgisayar oynamak istemediği için oyun sona erdi.")
                break
    
    # Oyun istatistikleri
    print(f"\nToplam Oynanan Oyun: {game_count}")
    print(f"Sizin Galibiyetleriniz: {player_wins}")
    print(f"Bilgisayarın Galibiyetleri: {computer_wins}")
    print("Görüşmek üzere!")

# Fonksiyonu çağırarak oyunu başlatabilirsiniz
tas_kagit_makas_RUKiYE_KARAATLI()
```
